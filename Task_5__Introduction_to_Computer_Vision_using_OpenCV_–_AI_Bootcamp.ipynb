## Section 3: Cropping and Rotation

### **Task 3**: Crop and Rotate the Image

*Instruction*: Use Keras to create a simple feedforward neural network with:


*   Crop the center square of the image
*   Rotate the image by 45 degrees using `cv2.getRotationMatrix2D`


```python
# Cropping
(h, w) = image.shape[:2]
center_crop = image[h//4:h*3//4, w//4:w*3//4]

# Rotation
M = cv2.getRotationMatrix2D((w//2, h//2), 45, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))

import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('sample.jpg')

# Get image dimensions
height, width = image.shape[:2]

# Step 1: Crop the center square of the image
min_dim = min(height, width)
start_x = (width - min_dim) // 2
start_y = (height - min_dim) // 2
cropped = image[start_y:start_y + min_dim, start_x:start_x + min_dim]

# Step 2: Rotate the cropped image by 45 degrees
center = (min_dim // 2, min_dim // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(cropped, rotation_matrix, (min_dim, min_dim))

# Step 3: Display results
plt.figure(figsize=(12, 4))

# Original
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')

# Cropped
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
plt.title('Center Cropped')
plt.axis('off')

# Rotated
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.title('Rotated 45')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Section 5: Filters and Edge Detection

### **Task 5**:  Apply Blur and Canny Edge Detection

*Instruction*: Apply Gaussian blur and detect edges using the Canny algorithm.


```python
blurred = cv2.GaussianBlur(image, (7, 7), 0)
edges = cv2.Canny(image, 100, 200)

import cv2
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('sample.jpg')

# Apply Gaussian Blur
blurred = cv2.GaussianBlur(image, (5, 5), 0)

# Apply Canny Edge Detection
edges = cv2.Canny(blurred, 50, 150)

# Display the original, blurred, and edge-detected images
plt.figure(figsize=(15, 5))

# Original image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Blurred image
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
plt.title('Gaussian Blurred')
plt.axis('off')

# Edge-detected image
plt.subplot(1, 3, 3)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edges')
plt.axis('off')

plt.tight_layout()
plt.show()
```
